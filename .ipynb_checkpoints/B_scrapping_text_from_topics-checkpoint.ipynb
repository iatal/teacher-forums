```python
df = pd.read_csv("/media/igna/Data/CRI_IA/Teacher_to_teacher/forum.csv")
```

```python
df['nb_pages'] = df['nb_pages'].replace(0,1)
```

```python
startTime = datetime.now()
stepTime = datetime.now()
print(startTime)

counter = 0
N = df['nb_pages'].sum()
NRows = len(df)
step = 1000
not_collected_urls = []

print("Parsing " + str(N) + " pages in " + str(math.ceil(NRows/step)) + " steps")

for i in range(math.ceil(NRows/step)):

    data = []
    
    for j in range(step):
        k = i*step + j
        if k >= NRows: break

        for pg_nb in range(df.iloc[k]["nb_pages"]):            
            url = df.iloc[k]["url_topic"] + "?page=" + str(pg_nb+1)
            
            forum_html_raw = simple_get(url)
            if forum_html_raw is None:
                not_collected_urls.append({'url' : url, 'position_in_df' : k, 'pg_nb' : pg_nb+1})
                print("Page not collected: " + url)
            else:    
                forum_html = BeautifulSoup(forum_html_raw)

                data_page = {}
                data_page['forum_name'] = df.iloc[k]["forum_name"]
                data_page['topic_name'] = df.iloc[k]["topic_name"]
                data_page['url_page'] = url

                time.sleep(0.25)
                data_topic = data_topic_extract(forum_html)
                data_page['data_topic'] = data_topic

                data.append(data_page)
    
    with open('/media/igna/Data/CRI_IA/Teacher_to_teacher/data_' + str(i+1) + '.json', 'w') as outfile:
        json.dump(data,outfile)
    
    print("Step " + str(i+1) + " completed in: " + str(datetime.now() - stepTime))
    stepTime = datetime.now()

print("All finished in " + str(datetime.now() - startTime))

print("Not collected pages: " + str(len(not_collected_urls)))

with open('/media/igna/Data/CRI_IA/Teacher_to_teacher/not_collected_urls.json', 'w') as outfile:
    json.dump(not_collected_urls,outfile)
```
