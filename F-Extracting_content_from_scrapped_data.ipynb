```python
path = '/media/igna/Data/CRI_IA/Teacher_to_teacher/'
fls = list(filter(lambda x:'data_' in x, os.listdir(path)))

all_comments = []

for one_file in fls:

    with open(path + one_file) as json_data:
        d = json.load(json_data)

    for topic in d:
        for article in topic["data_topic"]:
            for item in article["comment_text"]:
                all_comments.append(item)
```

```python
l_words = [word_tokenize(re.sub('[^A-Za-z0-9âàçëéêèïîôûù\-]+', ' ',sentence.lower())) for sentence in all_comments]
words = [item for sublist in l_words for item in sublist]
#l_words = []
#for wds in words: l_words.append(tokenizer.tokenize(wds))
#words = [item for sublist in l_words for item in sublist]
```

```python
len(words)
```

```python
filtered_words = [word for word in words if word not in stopwords.words('french')]
```

```python
stemmer = FrenchStemmer()
```

```python
stemmed_list = []
for x in filtered_words: stemmed_list.append(stemmer.stem(x))
```

```python
d = defaultdict(int)
for word in stemmed_list:
    d[word] += 1
```

```python
d_sorted = sorted(d.items(), key=lambda kv: kv[1],reverse=True)
```
